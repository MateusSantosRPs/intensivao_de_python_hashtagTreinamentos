# Automação de Sistemas e Processos com Python

### Desafio:

Todos os dias, o nosso sistema atualiza as vendas do dia anterior.
O seu trabalho diário, como analista, é enviar um e-mail para a diretoria, assim que começar a trabalhar, com o faturamento e a quantidade de produtos vendidos no dia anterior

E-mail da diretoria: seugmail+diretoria@gmail.com<br>
Local onde o sistema disponibiliza as vendas do dia anterior: https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing

Para resolver isso, vamos usar o pyautogui, uma biblioteca de automação de comandos do mouse e do teclado

#### Referencias para consulta 
 - https://pyautogui.readthedocs.io/en/latest/
 - https://pyperclip.readthedocs.io/en/latest/
 - https://docs.python.org/3/library/time.html
 - https://pandas.pydata.org/pandas-docs/stable/getting_started/index.html


```python
# bibliotecas para importar
import pyautogui
import time
import pyperclip

# Passo a Passo para resolver o desafio
pyautogui.PAUSE = 1 #entre uma linha de cod e outro espera 1sec

# 1 entrar no sistema da empresa (link Gdrive)

# Abrir navegador
pyautogui.hotkey("ctrl", "t") #abre nova aba no navegador
pyperclip.copy("https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing") #copiando o caminho do sistema
pyautogui.hotkey("ctrl", "v") #cola o link no navegador
pyautogui.press("enter") #entrando no sistema
time.sleep(12) #espera um tempo para carregar o sistema por ...(minha internet)


# 2 navegar no sistema (ate a pasta exportar)
pyautogui.click(x=320, y=255, clicks=2) #navega ate o diretorio e entra na pasta
time.sleep(5)

# 3 fazer o download da base de vendas
pyautogui.click(x=348, y=259) #seleciona o arquivo 
pyautogui.click(x=1159, y=163) #clica em tres pontinhos
pyautogui.click(x=945, y=562) #clica em download

time.sleep(5) #espera o tempo do download

```

### Vamos agora ler o arquivo baixado para pegar os indicadores

- Faturamento
- Quantidade de Produtos

```python
# 4 importar a base de vendas para o python
import pandas as pd

tabela = pd.read_excel(r"C:\Users\MATHE\Downloads\Vendas - Dez.xlsx") #, sheets=1,2...) caso tenha mais de uma aba no arquivo 
#tabela recebe(=) valor do arquivo excel (r sempre para caminhos)

display(tabela) #exibe a tabela (igual a um print)

# 5 colocar o faturamento e quantidade de produtos vendidos( indicadores)
faturamento = tabela["Valor Final"].sum()
quantidade = tabela["Quantidade"].sum()
```

### Vamos agora enviar um e-mail pelo gmail

```python
# 6 enviar email para a diretoria

pyautogui.hotkey("ctrl", "t")
pyperclip.copy("https://mail.google.com/mail/u/0/#inbox")
pyautogui.hotkey("ctrl", "v")
pyautogui.press("enter")
time.sleep(12)

# Passo 6: Enviar por e-mail o resultado
pyautogui.click(x=91, y=174) #clica no botão escrever novo email
time.sleep(10)

# Preencher o destino
pyautogui.write("mateus.contatos06@gmail.com") #escreve o destino do email
time.sleep(5)
pyautogui.press("tab") # seleciona o email
# escreve outro email
# tab
# escreve outro email
# tab
pyautogui.press("tab") # pula pro campo de assunto

# Preencher o assunto
pyperclip.copy("Relatório de Vendas")
pyautogui.hotkey("ctrl", "v") # escrever o assunto
pyautogui.press("tab") #pular pro corpo do email

# corpo do email variavel texto armazena a msg
texto = f"""
Prezados, bom dia

O faturamento de ontem foi de: R${faturamento:,.2f} 
A quantidade de produtos foi de: {quantidade:,}

Abs
Python"""

pyperclip.copy(texto)
pyautogui.hotkey("ctrl", "v")

# clicar no botão enviar
# apertar Ctrl Enter
pyautogui.hotkey("ctrl", "enter")
```
